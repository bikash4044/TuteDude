```python
import numpy as np
arr = np.array([i for i in range(1,21)])
print("Array: ", arr)
print("Sum: ", np.sum(arr))
print("Mean: ", np.mean(arr))
print("Median: ", np.median(arr))
print("Standard Deviation: ", np.std(arr))


```

Explanation:
- List comprehension is used to create the list of numbers
- numpy functions are used to compute sum, mean, median and standard deviation

```python
print("Indices of the element greater than 10 are:")
print([i for i in range(len(arr)) if arr[i]>10])
```

Explanation:
- indices of elements greater than 10 are fetched using list comprehension and then printed

```python
arr = np.array([i for i in range(1,17)])
arr2 = np.reshape(arr,(4,4))
print(arr2)
```

Explanation:
- list created using list comprehension and then converted into numpy array.
- array reshaped using numpy function

```python
print(arr2.T)
```

Explanation:
- np.array.T is used to create transpose of numpy array matrix

```python
print("Row-wise Sum: ", np.sum(arr2, axis=1))
print("Column-wise Sum: ", np.sum(arr2, axis=0))
```

Explanation:
- np.sum is used along axis 0 to generate column wise sum and along axis 1 to generate row wise sum

```python
arr3 = np.random.randint(1, 21, size=(3, 3))
arr4 = np.random.randint(1, 21, size=(3, 3))

print("First array: \n", arr3)
print("Second array: \n", arr4)

print("Addition of arrays:\n",arr3+arr4)
print("Subtraction of arrays:\n",arr3-arr4)
print("Multiplication of arrays:\n",arr3*arr4)
print("Dot product is same as Linear Multiplication.")
```

Explanation:
- np.random.randint generates random integers within given range and given shape
- mathematical binary operators are used to carry out linear operations

```python
print("Cross product of arrays:\n",np.dot(arr3,arr4))
```

Explanation:
- np.dot function is used to generate the cross product of matrix. In matrix the dot product is same as linear multiplication.

```python
arr_1d = np.arange(12)
print("1d Array:\n",arr_1d)
reshaped = arr_1d.reshape(3, 4)
print("Reshaped Array:\n",reshaped)
result = reshaped[:2, -2:]
print("Sliced Array:\n",result)
```

Explanation:
- np.arange returns all elements in given range nas an array
- array is reshaped and sliced as said
