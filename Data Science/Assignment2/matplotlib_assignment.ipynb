```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import color_sequences
from pygments import highlight

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y)
plt.grid()
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Explanation:
The line chart is plotted with plt.plot

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['red','green','yellow', 'blue'])
plt.title('Bar Chart')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

Explanation:
- plt.bar creates bar graph

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if val == max(revenue) else 0 for val in revenue]
plt.pie(revenue, labels=regions, autopct = '%1.1f%%',explode = explode)
plt.title('Pie Chart')
plt.show()
```

Explanation:
- plt.pie creates pie chart
- explode is used to highlight the value of max revenue

```python
import numpy as np
l = np.random.randint(0,101,1000)
plt.hist(l)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

Explanation:
- np.random.randint creates random values from starting value to ending -1 with given sample size
- plt.hist creates histogram plot of given data
