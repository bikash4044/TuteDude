```python
import pandas as pd
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
df = pd.DataFrame.from_dict(data)
df.head(5)
```

Explanation:
- dataframe is created from given dictionary and first n rows are displayed using head function

```python
df.describe()
```

Explanation:
- pandas describe function gives statistical details for numeric columns

```python
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print(f"The average salary in HR is: {hr_avg_salary}")
```

Explanation:
- dataframe is subset for department equals to HR and for the salary column, mean is computed

```python
avg_salary_per_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_per_dept)
```

Explanation:

```python
df.sort_values(by=['Age'], ascending=True, inplace=True)
df.to_csv("Srted.csv")
```
